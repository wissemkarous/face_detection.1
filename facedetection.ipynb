```python
def imshow(img):
    plt.imshow(img, )
    plt.axis('off')
    plt.show()
```

```python
def rgb_read(path, show=False):
    bgr_img = cv2.imread(path)
    rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
    
    if show:
        imshow(rgb_img)
    return rgb_img
```
